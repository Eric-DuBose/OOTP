```python
def prep_batters(df):
    data = df.copy()
    data = data[(data['league_id'] == 145) & (data['split_id'] == 1) & (data['year'] >= (season - 3)) & (data['year'] < season) & (data['pa'] >= data.pa.quantile(0.7))]
    columns = ['player_id','ab', 'h', 'k', 'pa', 'pitches_seen', 'g', 'gs',
       'd', 't', 'hr', 'r', 'rbi', 'sb', 'cs', 'bb', 'ibb', 'gdp', 'sh', 'sf']
    return data[columns].groupby('player_id').mean().astype(int).reset_index()
```

```python
batters = prep_batters(batters)
```
